# Exploratory Data Analysis (EDA)
Dataset: Titanic Dataset
Source: https://www.kaggle.com/datasets/yasserh/titanic-dataset

This notebook performs basic EDA using Pandas, Matplotlib, Seaborn, and Plotly.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load dataset (Assuming file is named 'titanic.csv' in same folder)
df = pd.read_csv('titanic.csv')
df.head()
```

## Summary Statistics

```python
df.describe()
```

## Histograms for Numerical Columns

```python
df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

## Boxplots

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.xticks(rotation=45)
plt.show()
```

## Correlation Matrix

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Pairplot

```python
sns.pairplot(df.select_dtypes(include='number'))
plt.show()
```

## Plotly Interactive Visualization

```python
fig = px.scatter(df, x='Age', y='Fare', color='Survived', title='Fare vs Age')
fig.show()
```
